```python
# Load the Pandas libraries with alias 'pd' 
import pandas as pd 
# Read data from file 'filename.csv' 
# (in the same directory that your python process is based)
# Control delimiters, rows, column names with read_csv (see later) 
data = pd.read_excel("sharing/Post-survey_Challenge_November 23, 2020_08.38.xlsx","QualtricsSurveysMerged") 
# Preview the first 5 lines of the loaded data 
data.head()
```

```python
validUsers = []

import re
p = re.compile('(\d\d?)\.(.*)')
articleId = re.compile('(\d\d?)_Q.*')
questionId = re.compile('.*_Q.*_(\d)')

for index, row in data.iterrows():
    user = {}
    evaluation = {}
    
    for pinfo in columns_excel:
        
        fetchID = articleId.match(pinfo)
        if fetchID:
            article = int(fetchID.group(1))-2

            fetchedQ = questionId.match(pinfo)
            question = int(fetchedQ.group(1))

            if article not in evaluation:
                evaluation[article] = {}

            evaluation[article][legende[question]] = row[pinfo]
        else:
            if pinfo in general:
                user[general[pinfo]] = row[pinfo]
                #user[pinfo] = row[pinfo]
    
    user["evaluations"] = evaluation
    if user["valid"] == True:
        validUsers.append(user)
```

```python
#distribution of users over the conditions

freq = {}
userMapping = {}

for x in validUsers:
    engine = engineMapping[x["recommender"]]
    userMapping[int(x["user_id"])] = engine
    
    if engine not in freq:
        freq[engine] = 0
    freq[engine] += 1
    
display(freq)
```

```python
# Formatting the User Evaluations of the Eight Recommendations for SPSS (Section 4.4.6)
evaluationScores = {}
for x in validUsers:
    engine = x["recommender"]
    if engine not in evaluationScores:
        evaluationScores[engine] = {}
    
    evals = x['evaluations']
    for i in evals:
        evaluation = evals[i]
        
        print(engine,",",engineMapping[engine],",",
             evaluation["relevance"],",",
             evaluation["novelty"],",",
             evaluation["topicality"],",",
             evaluation["attractiveness"])
        
        for y in evaluation:
            y_value = evaluation[y]
            if y not in evaluationScores[engine]:
                evaluationScores[engine][y] = {}
            
            if y_value not in  evaluationScores[engine][y]:
                evaluationScores[engine][y][y_value] = 0
            else:
                evaluationScores[engine][y][y_value]+=1
```

```python
import json

filename = 'interactionLogs.json'

behaviours = {}
behavioursArray = []
filehandle = open(filename, 'r')
while True:
    # read a single line
    line = filehandle.readline()
    if not line:
        break
    
    behaviour = json.loads(line)
    behaviour['user_id'] = int(behaviour['user_id'])
    behaviour["article"]["rec_pos"] = int(behaviour["article"]["rec_pos"])
    if behaviour['user_id'] in userMapping:
        behavioursArray.append(behaviour)
```

```python
import numpy as np

def precision_at_k(actual, predicted, k = 1):
    if len(predicted) > k:
        predicted = predicted[:k]
    relevant_items = np.isin(predicted, actual)
    precision = np.sum(relevant_items) / k
    return precision

def ndcg_at_k(actual, predicted, k = 1):
    if len(predicted) > k:
        predicted = predicted[:k]
    
    relevancy = np.zeros(k)
    if actual in predicted:
        pos = predicted.index(actual)
        relevancy[pos] = 1.0
        
    ideal_order =  sorted(relevancy,reverse=True)
    
    dcg = 0
    idcg = 0
    for i in range(k):
        
        dcg += relevancy[i] / np.log2(i + 2)
        idcg += ideal_order[i] / np.log2(i + 2)
        
    ndcg = dcg / idcg
    return ndcg
```

```python
# Formatting the User interactions for SPSS(Section 4.4.5)

for x in behavioursArray:
    uid = x["user_id"]
    
    aid = x["article"]["articleId"]
    recs = x["article"]["recommendation_list"]

    pos = -1
    if aid in recs:
        pos = recs.index(aid)
    if pos>=0:
        print(userMapping[uid],",", ndcg_at_k(aid, recs, 10),",", precision_at_k(aid,recs,10))
    else:
        print(userMapping[uid],",", 0.0,",",0)
```

```python
#clicks on articles and recommendation (bottom of Section 4.4.5)

readsAndRecs = {}
for x in behavioursArray:
    engine = userMapping[x["user_id"]]
    if engine not in readsAndRecs:
        readsAndRecs[engine] = {"articles_read":0,"recommendations_read":0}
        
    if x["article"]["recommendation"]:
        readsAndRecs[engine]["recommendations_read"]+=1
    
    readsAndRecs[engine]["articles_read"]+=1

print("users_survey:",len(userMapping),"users_interactions:",len(behaviours), "#interactions:",len(behavioursArray))
display(readsAndRecs)
```
